### Save RGB and Load RGB vector with reshape

```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import resize

def rgb2gray(rgb):
    if len(rgb.shape) is 3:
        return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
    else:
        # print ("Current Image if GRAY!")
        return rgb

img = imageio.imread("../DB/flower_photos/daisy/21652746_cc379e0eea_m.jpg")
grayimg = rgb2gray(img)
imgsize = [255, 255]
resizeImg = resize(img, [imgsize[0], imgsize[1]])
print(resizeImg.shape)

resizeGrayImg1 = resize(grayimg, [imgsize[0], imgsize[1]])
print(resizeGrayImg1.shape)

resizeGrayImg2 = resize(grayimg, [imgsize[0], imgsize[1]])/255.
plt.imshow(resizeGrayImg2)
plt.show()


```
